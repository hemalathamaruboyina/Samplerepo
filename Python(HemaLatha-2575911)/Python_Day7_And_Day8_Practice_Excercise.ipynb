# 2. Write a python program to implement binary search to find the target values from the list:
 1) Create a separate function to do binary search
 2) Get the number of inputs from the user
 3) Store the inputs individually in a list
 4) In binary search function at first sort the list 
 in order to start the search from the middle of the list
 5) Compare the middle element to right and left elements to  search targwt element
 6) If greater, move to right of list or else move to another side of the list
 7) Print the result along with the position of the element

```python
def binary_search(arr, key):
    arr.sort()
    l = 0
    r = len(arr) - 1
    while l <= r:
        mid = (l + r) // 2
        if arr[mid] == key:
            return mid  
        elif arr[mid] < key:
            l = mid + 1  
        else:
            r = mid - 1  
    return -1  
n = int(input("Enter the number of elements: "))
input_list = []
for i in range(n):
    element = int(input(f"Enter element {i+1}: "))
    input_list.append(element)
key = int(input("Enter the key element to search: "))
result = binary_search(input_list, key)
sorted_list = sorted(input_list)
print("Sorted List:", sorted_list)
if result != -1:
    print(f"Key element {key} is found at index {result}")
else:
    print(f"Key element {key} not found in the list")
```

# 6. Write a python script to perform the following operations on a singly linked list
1) Create a List
2) Find the smallest element from the list
3) Insert an element if it is not a duplicate element
4) Display the elements in reverse order

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
class SinglyLinkedList:
    def __init__(self):
        self.head = None
    def create_list(self):
        n = int(input("Enter the size of the list : "))
        for i in range(n):
            data = int(input(f"Enter {i+1} element: "))
            if not self.contains(data):
                new_node = Node(data)
                new_node.next = self.head
                self.head = new_node
    def contains(self, data):
        current = self.head
        while current:
            if current.data == data:
                return True
            current = current.next
        return False
    def find_smallest_element(self):
        if self.head is None:
            return None
        current = self.head
        smallest = current.data
        while current:
            if current.data < smallest:
                smallest = current.data
            current = current.next
        return smallest
    def display_reverse(self):
        self._display_reverse(self.head)

    def _display_reverse(self, current):
        if current is None:
            return
        else:
            self._display_reverse(current.next)
            print(current.data, end=" ")

if __name__ == "__main__":
    linked_list = SinglyLinkedList()
    while True:
        print("\nOptions:")
        print("1. Create a List")
        print("2. Find the Smallest Element")
        print("3. Insert an Element (if not duplicate)")
        print("4. Display Elements in Reverse Order")
        print("5. Reverse the List")
        print("6. Exit")
        choice = int(input("Enter your choice: "))
        if choice == 1:
            linked_list.create_list()
        elif choice == 2:
            smallest = linked_list.find_smallest_element()
            print(f"The smallest element in the list is {smallest}." if smallest is not None else "List is empty.")
        elif choice == 3:
            data = int(input("Enter the element to insert: "))
            if not linked_list.contains(data):
                new_node = Node(data)
                new_node.next = linked_list.head
                linked_list.head = new_node
                print(f"{data} inserted successfully.")
            else:
                print(f"{data} is already in the list (duplicate), not inserted.")
        elif choice == 4:
            linked_list.display_reverse()
        elif choice == 5:
            prev = None
            current = linked_list.head
            while current:
                next_node = current.next
                current.next = prev
                prev = current
                current = next_node
            linked_list.head = prev
            print("List reversed.")
        elif choice == 6:
            break
        else:
            print("Invalid choice. Please select a valid option.")
```

# 7. Write a python program to implement the various operations for Stack ADT
1) Push
2) Pop
3) Display

```python
class Stack:
    def __init__(self):
        self.elements = []
    def push(self, item):
        self.elements.append(item)
    def pop(self):
        if self.size() != 0:
            return self.elements.pop()
        else:
            raise IndexError("Pop from an empty stack")
    def size(self):
        return len(self.elements)  
    def display(self):
        if self.size() != 0:
            print("Stack elements:")
            for ele in reversed(self.elements):
                print(ele)
        else:
            print("Stack is empty")
if __name__ == "__main__":
    stk = Stack()
    while True:
        print("\nStack Operations:")
        print("1. Push")
        print("2. Pop")
        print("3. Display")
        print("4. Exit")
        choice = input("Enter your choice: ")
        if choice == '1':
            ele = input("Enter element to push: ")
            stk.push(ele)
            print(f"{ele} pushed onto the stack.")
        elif choice == '2':
            try:
                ele = stk.pop()  
                print(f"Popped element: {ele}")
            except IndexError as e:
                print(f"Error: {e}")
        elif choice == '3':
            stk.display()
        elif choice == '4':
            break
        else:
            print("Invalid choice. Please enter a valid option.")
```

# 3. Write a python program for sorting a list of elements using selection sort algorithm
1) Assume two lists: Sorted list - intially empty and unsorted list- given input list
2) In the first iteration, find the smallest element in the unsorted list and place it in the sorted list
3) In the second iteration, find the smallest element in the unsorted list 
and place it in the correct position by compariong with the element in the sorted lisst
4) In the third iteration, again find the smallest element in the unsorted list 
and place it in the correct position by comparing till the unsorted list becomes empty
5) This process continues till the unsorted list becomes empty
6) Display the sorted list

```python
def selection_sort(nums, n):
    for i in range(n - 1):
        ind = i
        for j in range(i + 1, n):
            if nums[j] < nums[ind]:
                ind = j
        nums[i], nums[ind] = nums[ind], nums[i]
nums= []
n = int(input("Enter the size of the list: "))
for i in range(n):
    ele =int(input(f"\nEnter the {i+1} element: "))
    nums.append(ele)
print(f"\nUnsorted list is\t: {nums}")
selection_sort(nums, n)
print(f"\nSorted list is\t\t: {nums}")
```

# 4. Write a pytho program for sorting a list of elements using insertion sort algorithm:
1) Assume two lists: Sorted list - Initially empty and Unsorted list - Given input list 
2) In the first Iteration, take the first element in the unsorted list and insert it in sorted list
3) In the second iteration, take the second element in the given list 
and compare with the element in the sorted sub list and place it in the correct position
4) In the third iteration, take the third element in the given list 
and compare with the elements in the sorted sub list and place the elements in the correct position
5) This process continures untill the last element is inserted in the sorted sublist
6) Display the sorted elements

```python
def insertion_sort(nums):
    for i in range(1, len(nums)):
        value = nums[i]  
        j = i - 1 
        while ((value < nums[j]) and (j>=0)):
            nums[j + 1] = nums[j]
            j -= 1
        nums[j + 1] = value
n = int(input("Enter the size of the list: "))
nums = []
for i in range(n):
    ele = int(input(f"Enter the {i+1} element: "))
    nums.append(ele)
print(f"Unsorted List is:\t{nums}")
insertion_sort(nums)
print(f"Sorted list is:\t\t{nums}")
```

# 5. Write a python program that performs merge sort on a list of numbers:
1) Divide : If the given array has zero or one element, return.
    i) Otherwise
    ii)Divide the input list in to two halves each containing half os the elements i.e. left half and right half
2) Conquer: Recursively sort the two sub-arrays (left half and right half)
    i) Call the merge sort on left half
    ii)Call the merge sort on right half
3) Combine: Combine the elements back in the input list by merging the two sorted lists into a sorted sequnce.

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr  
    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]
    left_half = merge_sort(left_half)
    right_half = merge_sort(right_half)
    return merge(left_half, right_half)
def merge(left, right):
    result = []
    left_index, right_index = 0, 0
    while left_index < len(left) and right_index < len(right):
        if left[left_index] < right[right_index]:
            result.append(left[left_index])
            left_index += 1
        else:
            result.append(right[right_index])
            right_index += 1
    result.extend(left[left_index:])
    result.extend(right[right_index:])
    return result
input_list = [int(x) for x in input("Enter a list of numbers separated by comma: ").split(",")]
sorted_list = merge_sort(input_list)
print("Sorted List:", sorted_list)
```

# 8. Write a python script to implement the various operations for Queue ADT
1) Insert
2) Delete
3) Display

```python
class Queue:
    def __init__(self):
        self.elements = []

    def insert(self, ele):
        self.elements.append(ele)

    def delete(self):
        if self.size() != 0:
            return self.elements.pop(0)
        else:
            raise IndexError("Delete from an empty queue")

    def size(self):
        return len(self.elements)

    def display(self):
        if self.size() != 0:
            print("Queue elements:")
            for ele in self.elements:
                print(ele)
        else:
            print("Queue is empty")



if __name__ == "__main__":
    queue = Queue()

    while True:
        print("\nQueue Operations:")
        print("1. Insert")
        print("2. Delete")
        print("3. Display")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            ele = input("Enter element to insert: ")
            queue.insert(ele)
            print(f"{ele} inserted into the queue.")
        elif choice == '2':
            try:
                ele = queue.delete()
                print(f"Deleted element: {ele}")
            except IndexError as e:
                print(f"Error: {e}")
        elif choice == '3':
            queue.display()
        elif choice == '4':
            break
        else:
            print("Invalid choice. Please enter a valid option.")
```

# 9. Write a python program to concert the following infix expressions to its postfix from using and pop operations of a stack
1) A/B^C+D*E-F*G
2) (B^2-4*A*C)^(1/2) (100)

```python
def infix_to_postfix(expression):
    def precedence(operator):
        precedence_dict = {"+": 1, "-": 1, "*": 2, "/": 2, "^": 3}
        return precedence_dict.get(operator, 0)
    def is_operator(char):
        return char in "+-*/^"
    def higher_precedence(op1, op2):
        return precedence(op1) > precedence(op2)
    postfix = []
    stack = []
    for char in expression:
        if char.isalnum():  
            postfix.append(char)
        elif char == "(":  
            stack.append(char)
        elif char == ")":  
            while stack and stack[-1] != "(":
                postfix.append(stack.pop())
            stack.pop()  
        elif is_operator(char):  
            while stack and stack[-1] != "(" and higher_precedence(stack[-1], char):
                postfix.append(stack.pop())
            stack.append(char)
    while stack:
        postfix.append(stack.pop())
    return "".join(postfix)
infix_expr1 = "A/B^C+D*E-F*G"
infix_expr2 = "(B^2-4*A*C)^(1/2) (100)"
postfix_expr1 = infix_to_postfix(infix_expr1)
postfix_expr2 = infix_to_postfix(infix_expr2)
print("Infix Expression 1\t:\t", infix_expr1)
print("Postfix Expression 1\t:\t", postfix_expr1)
print("\nInfix Expression 2\t:\t", infix_expr2)
print("Postfix Expression 2\t:\t", postfix_expr2)
```
