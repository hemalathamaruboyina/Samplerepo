# 1. Write a python program to create a base class "Shape"
 with methods to calculate area and perimeter.
 Then Create derived classes "Circle" and "Rectangle" 
 that inherit  from the base class and calculate their 
 respective area and perimeters.
 Demonstrate their usage in a program.

```python
import math
class Shape:
    def area(self):
        print("Area: ")
    def perimeter(self):
        print("Perimeter: ")
class Rectangle(Shape):
    def __init__(self, l, w):
        self.l = l
        self.w = w
    def area(self):
        return self.l * self.w
    def perimeter(self):
        return 2 * (self.l + self.w)
class Circle(Shape):
    def __init__(self, r):
        self.r = r
    def area(self):
        return math.pi * self.r ** 2
    def perimeter(self):
        return 2 * math.pi * self.r
if __name__ == "__main__":
    try:
        print("For Circle enter the radius\n")
        r = int(input("Enter radius : "))
        c = Circle(r)
        print("circle Radius is : ", c.r)
        print("circle Area is : ", c.area())
        print("Circle Perimeter is : ", c.perimeter())
        print("\nFor Rectangle enter the length and width\n")
        l = int(input("Enter the length of the rectangle: "))
        w = int(input("Enter the width of the rectangle: "))
        r = Rectangle(l, w)
        print("Rectangle of Length is : ", r.l)
        print("Rectangle of Width is : ", r.w)
        print("Rectangle of Area is : ", r.area())
        print("Rectangle of Perimeter is : ", r.perimeter())
        
    except BaseException as e:
        print(f"Exception: {e}")
```

# 2. You are developing an online quiz application
 where users can take quizzes on various topics and recive scores

 i) Create a class for quizzes and questions
 ii) Implement a scoring system that calculates the user's score on a quiz
 iii) How would you store and retrieve user progress, 
 Including quiz history and scores?

```python
class Questions:
    def __init__(self, text, options, crt_ans):
        self.text = text
        self.options = options
        self.crt_ans = crt_ans

class Quizzes:
    def __init__(self, title):
        self.title = title
        self.ques_list = []

    def add_question(self, question):
        self.ques_list.append(question)

    def calculate_score(self, user_ans):
        if len(user_ans) != len(self.ques_list):
            raise ValueError("Number of answers must match the number of questions in the quiz")
        score = 0
        for i, ans in enumerate(user_ans):
            if ans == self.ques_list[i].crt_ans:
                score += 1
        return score

class User:
    def __init__(self, username, register_number):
        self.username = username
        self.register_number = register_number
        self.quiz_history = {}

    def attempt_quiz(self, quiz, user_ans):
        score = quiz.calculate_score(user_ans)
        self.quiz_history[quiz.title] = score

    def get_quiz_score(self, quiz_title):
        return self.quiz_history.get(quiz_title, "Quiz not taken")

if __name__ == "__main__":
    name = input("Enter your name: ")
    roll_number = int(input("Enter your Roll number: "))
    user = User(name, roll_number)
    quiz = Quizzes("Python Quiz")
    print("\nPython QUIZ\n")

    # Define questions and add them to the quiz
    question1 = Questions("What is a correct syntax to output 'Hello World' in Python?",
                          ["echo 'Hello World'", "echo ('Hello World')", "print('Hello World')", "p('Hello World')"], 2)

    question2 = Questions("What is the output of the following code. p, q, r = 10, 20 ,30 print(p, q, r)?",
                          ["10 20", "10 20 30", "10", "Error: invalid syntax"], 1)

    question3 = Questions("What is the output of the following code.var= 'James Bond' print(var[2::-1])?",
                          ["Jam", "dno", "maJ", "dnoBsemaJ"], 2)

    question4 = Questions("What is the output of the following code?str = 'pynative' print (str[1:3])?",
                          ["py", "yn", "pyn", "yna"], 1)

    question5 = Questions("Which operator has higher precedence in the following list?",
                          ["% (Modulus)", "& (BitWise AND)", "** (Exponent)", "> (Comparison)"], 2)

    quiz.add_question(question1)
    quiz.add_question(question2)
    quiz.add_question(question3)
    quiz.add_question(question4)
    quiz.add_question(question5)

    user_ans = []

    for i, question in enumerate(quiz.ques_list, start=1):
        print(f"\nQuestion {i} : ")
        print(question.text)
        for j, option in enumerate(question.options, start=1):
            print(f"{j}: {option}")

        while True:
            try:
                ans = int(input(f"Enter option for Question {i}: "))
                if 1 <= ans <= len(question.options):
                    break
                else:
                    print("Invalid option. Please enter a valid option.")
            except ValueError:
                print("Invalid input. Please enter a valid option.")
        user_ans.append(ans - 1)

    user.attempt_quiz(quiz, user_ans)

    quiz_title = "Python Quiz"
    score = user.get_quiz_score(quiz_title)
    print(f"\n{user.username}'s score on the {quiz_title}: {score}/{len(quiz.ques_list)}")
```

# 3. Write a python script to create a class "Person"

```python
class Person:
    def __init__(self, name, age):
        self._name = name  
        self._age = age    
    def get_age(self):
        return self._age
    def is_eligible_to_vote(self):
        if self._age >= 18:
            return True
        else:
            return False
if __name__ == "__main__":
    name = input("Enter your name: ")
    age = int(input("Enter your age: "))
    person = Person(name, age)
    if person.is_eligible_to_vote():
        print(f"{person.get_age()}-year-old \n{person._name} is eligible to vote.")
    else:
        print(f"{person.get_age()}-year-old \n{person._name} is not eligible to vote.")
```

# 4. You are tasked with designing a python class hierarchy for a simple banking system.
 The system should be able to handle different types of accounts, 
 such as savings accounts and checking accounts.

 Both account types should have common attributes like an account number, account holders name and balance.
 However, savings accounts should have an additional attribute for interest rate 
 while checking accounts should have attribute for overdraft limit

 i) Create a python class called bank-account with the following attributes and methods:
     a) Attributes: Account_number, account_holder_name and balance
     b) Methods: __init__()(constructor), deposit(), and withdraw()
 ii) Create two subclasses, SavingsAccount and CheckingAccount, that inherit from the BankAccount class
 iii) Add the following attributes and methods to each subclass"
     a) SavingsAccount:
         i) Additional attribute: Interest_rate
         ii)Method: Calculate_interest(), which calculate and add interest to the account base in the interest rate
     b)CheckingAccount:
         i)Additional attribute: overdraft_limit
         ii)Method: withdraw allows withdrawung money up to the overdraft limit (if available) without additional fees
 iv)write a program that creates instances of both SavingsAccount and CheckingAccount 
 and demonstrates the use of their methods.
 v) Implement proper encapsulation by making the attributes private where necessary 
 and providing getter and setter methods as needed
 vi) Handle any potential errors or exceptions 
 that may occur during operations like withdrawls, deposits or interest calculations
 vii) Provide comments in your code to explain the purpose of each class, attribute and mehod

 Note: You code should create instances of the classes, simulate transactions 
 and showcase the differences between Savings  Accounts and Checking Accounts.

```python
class BankAccount:
    def __init__(self, account_number, account_holder_name, balance=0.0):
        self._account_number = account_number  
        self._account_holder_name = account_holder_name
        self._balance = balance 
    def get_account_number(self):
        return self._account_number
    def get_account_holder_name(self):
        return self._account_holder_name
    def get_balance(self):
        return self._balance
    def deposit(self, amount):
        try:
            if amount > 0:
                self._balance += amount
                return f"Deposit of ${amount} successful. New balance: ${self._balance}"
            else:
                raise ValueError("Amount to deposit must be greater than 0.")
        except ValueError as e:
            return str(e)
    def withdraw(self, amount):
        try:
            if amount > 0 and amount <= self._balance:
                self._balance -= amount
                return f"Withdrawal of ${amount} successful. New balance: ${self._balance}"
            else:
                raise ValueError("Invalid withdrawal amount.")
        except ValueError as e:
            return str(e)
class SavingsAccount(BankAccount):
    def __init__(self, account_number, account_holder_name, balance=0.0, interest_rate=0.0):
        super().__init__(account_number, account_holder_name, balance)
        self._interest_rate = interest_rate  
    def get_interest_rate(self):
        return self._interest_rate
    def calculate_interest(self):
        try:
            if self._interest_rate >= 0:
                interest = (self._interest_rate / 100) * self._balance
                self._balance += interest
                return f"Interest calculated and added. New balance: ${self._balance}"
            else:
                raise ValueError("Interest rate must be non-negative.")
        except ValueError as e:
            return str(e)
class CheckingAccount(BankAccount):
    def __init__(self, account_number, account_holder_name, balance=0.0, overdraft_limit=0.0):
        super().__init__(account_number, account_holder_name, balance)
        self._overdraft_limit = overdraft_limit  # Private attribute for overdraft limit
    def get_overdraft_limit(self):
        return self._overdraft_limit
    def withdraw(self, amount):
        try:
            if amount > 0:
                if amount <= (self._balance + self._overdraft_limit):
                    self._balance -= amount
                    return f"Withdrawal of ${amount} successful. New balance: ${self._balance}"
                else:
                    raise ValueError("Withdrawal amount exceeds available funds and overdraft limit.")
            else:
                raise ValueError("Invalid withdrawal amount.")
        except ValueError as e:
            return str(e)
if __name__ == "__main__":
    savings_account = SavingsAccount("SBI123", "Hema", balance=1000.0, interest_rate=5.0)
    print(f"Account holder: {savings_account.get_account_holder_name()}")
    print(f"Initial balance: ${savings_account.get_balance()}")
    print(savings_account.calculate_interest())
    print(savings_account.deposit(500))
    print(savings_account.withdraw(200))
    checking_account = CheckingAccount("HDFC456", "Latha", balance=500.0, overdraft_limit=200.0)
    print(f"\nAccount holder: {checking_account.get_account_holder_name()}")
    print(f"Initial balance: ${checking_account.get_balance()}")
    print(checking_account.deposit(100))
    print(checking_account.withdraw(700))
```

 iii) Override the salary calculation method in both subclasses to account for different payment structures.

```python
class Emp:
    def __init__(self, name, emp_id, salary):
        self.__name = name
        self.__emp_id = emp_id
        self.__salary = salary
    def get_name(self):
        return self.__name
    def set_name(self, name):
        self.__name = name
    def get_emp_id(self):
        return self.__emp_id
    def set_emp_id(self, emp_id):
        self.__emp_id = emp_id

    def get_salary(self):
        return self.__salary

    def set_salary(self, salary):
        self.__salary = salary

class FullTimeEmp(Emp):
    def __init__(self, name, emp_id, annual_salary):
        super().__init__(name, emp_id, annual_salary)

    def calculate_salary(self):
        return self.get_salary()

class PartTimeEmp(Emp):
    def __init__(self, name, emp_id, hours_worked, hourly_rate):
        super().__init__(name, emp_id, None)
        self.__hours_worked = hours_worked
        self.__hourly_rate = hourly_rate

    def calculate_salary(self):
        return self.__hours_worked * self.__hourly_rate
full_time_emp_name = input("Enter Full-Time Employee's Name: ")
full_time_emp_id = input("Enter Employee ID: ")
full_time_emp_salary = float(input("Enter Annual Salary: "))
part_time_emp_name = input("Enter Part-Time Employee's Name: ")
part_time_emp_id = input("Enter Employee ID: ")
part_time_emp_hours = float(input("Enter Hours Worked: "))
part_time_emp_hourly_rate = float(input("Enter Hourly Rate: "))
full_time_emp = FullTimeEmp(full_time_emp_name, full_time_emp_id, full_time_emp_salary)
part_time_emp = PartTimeEmp(part_time_emp_name, part_time_emp_id, part_time_emp_hours, part_time_emp_hourly_rate)
print("\nEmployee Information:")
print(f"{full_time_emp.get_name()} (ID: {full_time_emp.get_emp_id()}) Salary: ${full_time_emp.calculate_salary()}")
print(f"{part_time_emp.get_name()} (ID: {part_time_emp.get_emp_id()}) Salary: ${part_time_emp.calculate_salary()}")
```

# 6. library management system:
 Scnario: you are developing a library management system 
 where you need to handle books, partrons, and Library transactions.

 1: Create a class hierarchy that includes classes for books (e.g., Book), 
 patrons (e.g., Patron), and transactions (e.g., Transaction).
 Define attributes and methods for each class

 2: Implement encapsulation by making relevant attributes private 
 and providing getter and setter methods where necessary

 3: Use inheritance to represent different types of books (e.g., fiction, non-fiction) 
 as subclasses of the book class. Ensure that each book type can have specific attributes and methods

 4:Demonstrate polymorphism by allowing patrons to check out and return books, regardless of the book type.

 5: Implement a method for tracking overdue books and notifying patrons

 6: Consider scenarios like book reservations, late fees and library staff interactions in your design

```python
from datetime import datetime, timedelta
class Book:
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.checked_out = False
class FictionBook(Book):
    def __init__(self, title, author, isbn, genre):
        super().__init__(title, author, isbn)
        self.genre = genre
class NonFictionBook(Book):
    def __init__(self, title, author, isbn, topic):
        super().__init__(title, author, isbn)
        self.topic = topic
class Patron:
    def __init__(self, name, patron_id):
        self.name = name
        self.patron_id = patron_id
class Transaction:
    def __init__(self, book, patron, due_date):
        self.book = book
        self.patron = patron
        self.due_date = due_date

def checkout_book(patron, book, due_date):
    if not book.checked_out:
        book.checked_out = True
        return Transaction(book, patron, due_date)
    else:
        return None

def return_book(transaction):
    if transaction.book.checked_out:
        transaction.book.checked_out = False
def check_overdue_transactions(transactions):
    today = datetime.now()
    overdue_transactions = []

    for transaction in transactions:
        due_date = transaction.due_date
        if today > due_date:
            overdue_transactions.append(transaction)
    return overdue_transactions
def notify_patrons(overdue_transactions):
    for transaction in overdue_transactions:
        patron = transaction.patron
        print(f"Dear {patron.name}, your book '{transaction.book.title}' is overdue.")
def main():
    print("Enter Book Details:")
    book_title = input("Title: ")
    book_author = input("Author: ")
    book_isbn = input("ISBN: ")
    book_genre = input("Genre (if fiction, else leave empty): ")
    book_topic = input("Topic (if non-fiction, else leave empty): ")
    if book_genre:
        book = FictionBook(book_title, book_author, book_isbn, book_genre)
    else:
        book = NonFictionBook(book_title, book_author, book_isbn, book_topic)
    print("\nEnter Patron Details:")
    patron_name = input("Name: ")
    patron_id = input("Patron ID: ")

    patron = Patron(patron_name, patron_id)

    due_date = datetime.now() + timedelta(days=14)  # Due date is set to 14 days from today

    transaction = checkout_book(patron, book, due_date)

    if transaction:
        print(f"\nBook '{book.title}' checked out by {patron.name}. Due Date: {due_date.strftime('%Y-%m-%d')}")

if __name__ == "__main__":
    main()
```

# 7: Online Shopping Cart
 Scenario: You are tasked with designing a class hierarchy for an online shoping cart system.
 The system should handle products, shoping carts 
 and other orders. Consider various OOP principles while designing this system.

 i) Create a class hierarchy that includes classes for products (e.g., Order). 
 Define attributes and methods for each class

 ii) Implement encapsulation by makinf relevant attributes private 
 and providing getter and setter mehtods where necessary

 iii) Use inheritance to represent different types of products 
 (e.g., Elecrtonics, clothing, Food and bevarages, Healthcare) as subclass of the product class.
 Ensure that each product type can have specific attributes and methods

 iv) Demonstrate polymorphis by allowing various product types 
 to be added to a shoping cart and calculate the total cost of items in the cart.

 v) Implement a method for placing an order, which transfers items from the shoping cart to an order.

 Consider scenarios like out-of-stock products, discounts, and shipping costs in your design.

```python
class Product:
    def __init__(self, product_id, name, price):
        self.__product_id = product_id
        self.__name = name
        self.__price = price

    def get_product_id(self):
        return self.__product_id

    def get_name(self):
        return self.__name

    def set_name(self, name):
        self.__name = name

    def get_price(self):
        return self.__price

    def set_price(self, price):
        self.__price = price

class Electronics(Product):
    def __init__(self, product_id, name, price, brand):
        super().__init__(product_id, name, price)
        self.__brand = brand

    def get_brand(self):
        return self.__brand

    def set_brand(self, brand):
        self.__brand = brand

class Clothing(Product):
    def __init__(self, product_id, name, price, size, color):
        super().__init__(product_id, name, price)
        self.__size = size
        self.__color = color

    def get_size(self):
        return self.__size

    def set_size(self, size):
        self.__size = size

    def get_color(self):
        return self.__color

    def set_color(self, color):
        self.__color = color

class ShoppingCart:
    def __init__(self):
        self.__items = []

    def add_item(self, product, quantity=1):
        self.__items.append({"product": product, "quantity": quantity})

    def remove_item(self, product):
        for item in self.__items:
            if item["product"] == product:
                self.__items.remove(item)

    def get_items(self):
        return self.__items

    def calculate_total_cost(self):
        total_cost = sum(item["product"].get_price() * item["quantity"] for item in self.__items)
        return total_cost

class Order:
    def __init__(self, order_id):
        self.__order_id = order_id
        self.__customer_name = None
        self.__customer_phone = None
        self.__shipping_address = None
        self.__order_items = []

    def set_customer_details(self, name, phone, address):
        self.__customer_name = name
        self.__customer_phone = phone
        self.__shipping_address = address

    def add_order_item(self, product, quantity):
        self.__order_items.append({"product": product, "quantity": quantity})

    def calculate_order_total(self):
        total_cost = sum(item["product"].get_price() * item["quantity"] for item in self.__order_items)
        return total_cost

def calculate_shipping_cost(order_total):
    if order_total > 100:
        return 5
    else:
        return 10

def display_product_menu(products):
    print("Available Products:")
    for i, product in enumerate(products, start=1):
        print(f"{i}. {product.get_name()} - ${product.get_price()}")

def select_products(products):
    selected_products = []
    while True:
        display_product_menu(products)
        choice = input("Enter the product number to add to your cart (0 to finish): ")
        if choice == '0':
            break
        try:
            choice = int(choice)
            if 1 <= choice <= len(products):
                selected_product = products[choice - 1]
                quantity = int(input(f"How many {selected_product.get_name()} would you like to add? "))
                selected_products.append((selected_product, quantity))
            else:
                print("Invalid choice. Please select a valid product number.")
        except ValueError:
            print("Invalid input. Please enter a number.")
    return selected_products

if __name__ == "__main__":
    # Create some product instances
    products = [
        Electronics("P001", "iPhone 12", 999.99, "Apple"),
        Electronics("P002", "Samsung Galaxy S21", 899.99, "Samsung"),
        Clothing("P003", "Cotton T-Shirt", 19.99, "M", "Blue"),
        Clothing("P004", "Jeans", 39.99, "32", "Black"),
        Clothing("P005", "Sneakers", 49.99, "10", "White"),
    ]

    selected_products = select_products(products)

    cart = ShoppingCart()
    for product, quantity in selected_products:
        cart.add_item(product, quantity)

    order = Order("O001")

    customer_name = input("Enter customer name: ")
    customer_phone = input("Enter customer phone number: ")
    shipping_address = input("Enter shipping address: ")

    order.set_customer_details(customer_name, customer_phone, shipping_address)

    for item in cart.get_items():
        order.add_order_item(item["product"], item["quantity"])

    cart_total = cart.calculate_total_cost()

    discounted_total = order.calculate_order_total()
    discounted_total = discounted_total * 0.9  # Apply a 10% discount

    shipping_cost = calculate_shipping_cost(discounted_total)

    final_total = discounted_total + shipping_cost

    print(f"Order Total: ${cart_total}")
    print(f"Discounted Total: ${discounted_total}")
    print(f"Shipping Cost: ${shipping_cost}")
    print(f"Final Total: ${final_total}")
```
